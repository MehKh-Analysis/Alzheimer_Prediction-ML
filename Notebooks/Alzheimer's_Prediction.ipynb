```python
# using the 1.5*IQR rule to ensure there are no outliers
outlier_counts = {}

for col in colummns_continous:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    outlier_counts[col] = len(outliers)

pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier Count'])
```

## Building Classification Models :
 (**Logistic Regression, Naive Bayes, KNN, Random Forest, and SVM**)

 + ***Hyperparameter tuning using CV***

```python
# -------------------- 1. Imports --------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
)

# -------------------- 2. Data prep --------------------
top_5_features = [
    "FunctionalAssessment",
    "ADL",
    "MemoryComplaints",
    "MMSE",
    "BehavioralProblems",
]

X_train, X_test, y_train, y_test = train_test_split(
    X[top_5_features], y, test_size=0.30, stratify=y, random_state=42
)

# -------------------- 3. Model + grid factory --------------------
def model_registry():
    return {
        "LogReg": (
            LogisticRegression(max_iter=10_000, random_state=42),
            {"penalty": ["l1", "l2"], "C": [1e-2, 1e-1, 1, 10], "solver": ["saga"]},
        ),
        "NaiveBayes": (GaussianNB(), {}),
        "KNN": (KNeighborsClassifier(), {"n_neighbors": [3, 5, 7]}),
        "RandomForest": (
            RandomForestClassifier(random_state=42),
            {"n_estimators": [50, 100], "max_depth": [None, 5, 10]},
        ),
        "SVM": (
            SVC(random_state=42),
            {"C": [0.1, 1, 10], "kernel": ["linear", "rbf"]},
        ),
    }

# -------------------- 4. Grid-search helper --------------------
def run_grid(name, est, grid, X_tr, y_tr, cv):
    pipe = Pipeline(
        [("scaler", MinMaxScaler()), ("model", est)]
    )
    grid = {f"model__{k}": v for k, v in grid.items()}
    gs = GridSearchCV(
        pipe,
        grid,
        cv=cv,
        scoring="accuracy",
        n_jobs=-1,
        verbose=0,
        return_train_score=False,
    )
    gs.fit(X_tr, y_tr)
    print(f"{name:<12} best params: {gs.best_params_}")
    return gs.best_estimator_

# -------------------- 5. Train with CV --------------------
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
best_models = {
    name: run_grid(name, est, grid, X_train, y_train, cv)
    for name, (est, grid) in model_registry().items()
}

# -------------------- 6. Evaluate on test set --------------------
test_scores = {}
for name, model in best_models.items():
    y_pred = model.predict(X_test)
    test_scores[name] = accuracy_score(y_test, y_pred)

    print(f"\n=== {name} ===")
    print("Classification report:")
    print(classification_report(y_test, y_pred, digits=3, zero_division=0))

    cm = confusion_matrix(y_test, y_pred)
    cm_df = pd.DataFrame(
        cm,
        index=[f"True {c}" for c in np.unique(y)],
        columns=[f"Pred {c}" for c in np.unique(y)],
    )
    print("Confusion matrix:")
    print(cm_df)

# -------------------- 7. Bar-plot model comparison --------------------
# Sort test_scores by accuracy (ascending)
sorted_scores = dict(sorted(test_scores.items(), key=lambda item: item[1]))

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(sorted_scores.keys(), sorted_scores.values())
plt.ylabel("Accuracy (test set)")
plt.title("Model Comparison on Top-5 Features (Ascending)")
plt.ylim(0, 1)
plt.grid(axis="y", ls="--", alpha=0.5)
plt.xticks(rotation=45)  # Optional: rotate labels if needed
plt.tight_layout()
plt.show()

```

```python
# Balanced Random Forest: Modifies the standard Random Forest algorithm to address class imbalance
# by undersampling the majority class in each bootstrap sample.
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming X (your features) and y (your target) are already loaded

# Split data
X_train, X_test, y_train, y_test = train_test_split(X[top_5_features], y, test_size=0.3, random_state=42)

# Instantiate and train Balanced Random Forest
model = BalancedRandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred_balanced_rf = model.predict(X_test)

# Evaluate performance
accuracy_balanced_rf = accuracy_score(y_test, y_pred_balanced_rf)
cm_balanced_rf = confusion_matrix(y_test, y_pred_balanced_rf)
report_balanced_rf = classification_report(y_test, y_pred_balanced_rf, zero_division=0)

print("\n--- Balanced Random Forest Results ---")
print(f"Accuracy (Balanced RF): {accuracy_balanced_rf:.4f}")
print(f"Confusion Matrix (Balanced RF):\n{cm_balanced_rf}")
print(f"Classification Report (Balanced RF):\n{report_balanced_rf}")

plt.figure(figsize=(8, 6))
sns.heatmap(cm_balanced_rf, annot=True, fmt='d', cmap='Blues',
            xticklabels=sorted(pd.Series(y).unique()),
            yticklabels=sorted(pd.Series(y).unique()))
plt.title('Confusion Matrix - Balanced Random Forest')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

## Model Interpretebility ( SHAP )

```python
import shap
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- Assumes X_scaled_top_5_df and y are already defined ---

# âœ… Ensure y is clean and binary
y = pd.Series(y).astype(int).squeeze()
print(f"âœ… y shape: {y.shape}, unique values: {y.unique()}")

# âœ… Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X[top_5_features], y, test_size=0.3, random_state=42
)

# âœ… Train model
balanced_rf = BalancedRandomForestClassifier(n_estimators=100, random_state=42)
balanced_rf.fit(X_train, y_train)

# âœ… SHAP Explainer
explainer = shap.TreeExplainer(balanced_rf)

# âœ… Subset of test data for SHAP analysis
X_test_subset = X_test.iloc[:100].copy()
shap_values = explainer.shap_values(X_test_subset)

# âœ… Select SHAP values for class 1 (Alzheimer)
shap_class_1 = shap_values[:, :, 1]


print("\nðŸ” Generating SHAP Summary Plot...")
shap.summary_plot(
    shap_class_1,
    X_test_subset,
    feature_names=X_test_subset.columns
)
# ðŸ”Ž Print mean |SHAP| value per feature (class-1 influence)
feature_imp = pd.Series(np.abs(shap_class_1).mean(axis=0), index=X_test_subset.columns)
print("\nMean |SHAP| per feature (class 1):\n", feature_imp.sort_values(ascending=False))
```
